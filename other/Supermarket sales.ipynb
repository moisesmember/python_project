<h3>Detalhes financeiros</h3>

<h4>Faturamento por Categoria (Branch)</h4>

```python
agrupar_branch = sales.groupby(['Branch','City']).agg({'Total':sum})
lucro_categoria = agrupar_branch.apply(lambda x: x.sort_values(ascending=False).head(4))
lucro_categoria
```

<h4>Quantidade de Vendas por Categoria (Branch)</h4>

```python
qtde_venda = sales.groupby('Branch')['Branch'].count().sort_values(ascending=False)
qtde_venda
```

<h4>Quantidade de Vendas por Produto</h4>

```python
# Lucro de Produtos por Categoria
agrupar_product = sales.groupby(['Product_line','Branch','City']).agg({'Total':sum})
lucro_produto_categoria = agrupar_product.apply(lambda x: x.sort_values(ascending=False).head(10))
lucro_produto_categoria
```

```python
# Lucro por Produtos
agrupar_product = sales.groupby(['Product_line']).agg({'Total':sum})
lucro_produto = agrupar_product.apply(lambda x: x.sort_values(ascending=False).head(10))
lucro_produto
```

```python
# Quantidade de Venda por Produto
qtde_venda_produto = sales.groupby('Product_line')['Product_line'].count().sort_values(ascending=False)
# O valor retornado é uma pandas.Series
# qtde_venda_produto.index
# qtde_venda_produto.values
# qtde_venda_produto.size

# Conversão de pandas.Series para Dataframe
qtde_venda_produtos = pd.DataFrame({'product':qtde_venda_produto.index, 'qtde':qtde_venda_produto.values})
qtde_venda_produtos
```
