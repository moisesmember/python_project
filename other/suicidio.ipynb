```python
print('====== 1 quartil =======')
print(dataset_work.quantile(0.25)) # quartil 0 <= x <= 1
print('====== 2 quartil =======')
print(dataset_work.quantile(0.5)) # quartil 0 <= x <= 1
print('====== 3 quartil =======')
print(dataset_work.quantile(0.75)) # quartil 0 <= x <= 1
print('====== 4 quartil =======')
print(dataset_work.quantile(1)) # quartil 0 <= x <= 1

print('------------------------')
print('Mediana: ',dataset_work.median()) # O 2 QUARTIL É IGUAL A MEDIANA
```

```python
# dataset_work.var() # variância
dataset_work.std()
```

```python
skewness = dataset_work.skew() # mede o grau de simetria da curva que deve é bem próxima de zero

if skewness['suicides_no'] > 0:
    print('Positive Skew - {moda < mediana < média}')
elif skewness['suicides_no'] < 0:
    print('Negative Skew - {média < mediana < moda}')
elif skewness['suicides_no'] == 0:
    print('Symmetrical Distribution - {moda = mediana = média}')
```
